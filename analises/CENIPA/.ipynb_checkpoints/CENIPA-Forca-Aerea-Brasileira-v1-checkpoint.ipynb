```python
fatores = pd.read_csv('./data/fator_contribuinte.csv', sep=';', skip_blank_lines=True)
fatores = fatores.loc[:,['codigo_ocorrencia3','fator_nome','fator_area']]
```

```python
# Os acidentes envolvem fatores OPERACIONAL, HUMANA ou MATERIAL, e de outra(OUTRO) natureza?
# FATOR que tem area *** será classificado como OUTRO.
# seleciona somente as linhas que contem '***' na coluna fator_area, e troca por 'OUTRO'
fatores.loc[fatores.fator_area == '***','fator_area'] = 'OUTRO'
# 159,1654 verificando os indices que foram trocados.
fatores.loc[[159,1654],'fator_area']
# grafico de colunas com o numero de ocorrencia que foram classificados por area de fotores.
nd = fatores.groupby('fator_area').count().sort_values(by='codigo_ocorrencia3',ascending=False)
nd.plot.bar(y='codigo_ocorrencia3',ylabel='Número de Ocorrências',xlabel='Área do Fator',title='Número de Ocorrências por Area')
```

```python
# Quais os principais fatores que contribuem para a ocorrencia de acidentes com aeronaves considerando apenas
# fatores operacionais com incidencia maiores que 10?
pp = fatores.loc[fatores.fator_area == 'FATOR OPERACIONAL']#.sort_values(by='fator_area')#,ascending=False)
gby = pp.groupby('fator_nome').count().sort_values(by=['codigo_ocorrencia3'],ascending=True)
gby10 = gby[gby.codigo_ocorrencia3 > 10].div(len(pp))

gby10.mul(100).plot.barh(y='codigo_ocorrencia3')
gby10.mul(100).tail(6)
# O fator MANUTENÇÃO DA AERONAVE é o quinto fator que tem infuÊncia no acidedentes com aeronaves (9,1%). Antecedido por:
# em primeiro lugar o JULGAMENTO DE PILOTAGEM, em segundo APLICAÇÃO DE COMANDOS, em terceiro SUPERVISÃO GERENCIAL, em
# quarto PLANEJAMENTO DE VOO.
```

```python
# Quais os principais fatores que contribuem para a ocorrencia de acidentes com aeronaves considerando apenas
# fatores humanos com incidencia maiores que 10?
pp = fatores.loc[fatores.fator_area == 'FATOR HUMANO']#.sort_values(by='fator_area')#,ascending=False)
gby = pp.groupby('fator_nome').count().sort_values(by=['codigo_ocorrencia3'],ascending=True)
gby10 = gby[gby.codigo_ocorrencia3 > 10].div(len(pp))

gby10.mul(100).plot.barh(y='codigo_ocorrencia3')
gby10.mul(100).tail(6)
# O fator CAPACITAÇÃO E TREINAMENTO é o sexto fator que tem infuÊncia no acidedentes com aeronaves. E está relatado
# em 5,8% dos casos
```

Quais fatores mais influenciaram para o acidente acontecer? netes 10 anos dos acidentes registrados.

```python
# Estre as areas dos fatores quais contribuem para a ocorrencia de acidentes com aeronaves com incidencia maiores que 10?
ppT = fatores.groupby(['fator_nome']).count().sort_values(by=['fator_area','fator_nome'],ascending=True)
g1 = ppT[ppT.fator_area > 50].div(len(fatores))
g1.tail(10).mul(100)
g1.mul(100).plot.barh(y='codigo_ocorrencia3')
# ok
```

### Total de Incidentes e Acidentes

```python
# ocorrencia['ocorrencia_classificacao'].value_counts().sort_values(ascending=False).div(len(ocorrencia)).mul(100)
datas = ocorrencia['ocorrencia_classificacao'].value_counts().sort_values(ascending=False).div(len(ocorrencia)).mul(100)
datas
figure = plt.figure(figsize=[12,10])
plt.bar(datas.index,datas.values)

for i in range(len(datas)):
    plt.text(datas.index[i], datas.values[i]+1.5,f'{datas.values[i]:.1f}',size=12)

plt.xlabel('Classificação das ocorrências',size=12)
plt.ylabel('Porcentagem das ocorrências',size=12)
plt.ylim(0,100)
plt.show()
figure.savefig('classificacao_ocor_porcent.png', bbox_inches='tight' )
```

```python
np.array(datas.values).sum()
```
